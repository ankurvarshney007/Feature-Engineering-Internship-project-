```python
from scipy.spatial.distance import mahalanobis
import scipy as sp
import pandas as pd


df = df.iloc[:,5:8]

Sx = df.cov().values
Sx = sp.linalg.inv(Sx)

mean = df.mean().values

def mahalanobisR(X,meanCol,IC):
    m = []
    for i in range(X.shape[0]):
        m.append(mahalanobis(X.iloc[i,:],meanCol,IC) ** 2)
    return(m)

mR = mahalanobisR(df,mean,Sx)
```

```python
df['distance'] = mR
```

```python
df.to_excel("mahanabolis11.xlsx", index=False)
```
